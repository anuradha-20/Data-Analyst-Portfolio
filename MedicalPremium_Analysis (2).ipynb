```python
df1 = df[df["NumberOfMajorSurgeries"]  == 2]          # counting Premium Price for users who have had 2 surgeries done.
df1['PremiumPrice'].value_counts()
```

```python
plt.figure(figsize=(15,5))
(df1['PremiumPrice'].value_counts(normalize=True)).plot(kind='pie',explode=[0,0.1,0.1,0.1])
plt.show()
```

```python
df1 = df[df["NumberOfMajorSurgeries"]  == 3]
df1['PremiumPrice'].value_counts()
```

# From the above observation we can clearly say that customer who have done more than two major surgeries pay a standard premium of 28000rs.
